### summarizing data

```python
# analogous to methods used with numpy earlier

print(difdf.Correct[difdf.Condition == 1].mean())
print(difdf.Correct[difdf.Condition == 2].mean())
```

### pivot table

```python
# https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.pivot_table.html

pt = pd.pivot_table(difdf, values='Correct', index='Condition')
pt
```

```python
print(pt)
print(type(pt))
print(pt.iloc[0, 0])
print(pt.iloc[1, 0])
```

```python
pd.pivot_table(difdf, values='RT', index='Condition')
```

```python
minRT = .1
maxRT = 1.

pd.pivot_table(difdf[(difdf.RT>minRT) & (difdf.RT<maxRT)], values='Correct', index='Condition')
```

```python
# pivot table calculates means 

pd.pivot_table(difdf[(difdf.RT>minRT) & (difdf.RT<maxRT)], values='RT', index='Condition')
```
